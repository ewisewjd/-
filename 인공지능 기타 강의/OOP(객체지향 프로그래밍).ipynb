# oop (object oriented programming)
- 객체 지향 프로그래밍
    - 집을 짓는다면 집은 객체이고 집을 만들기 위한 설계도는 class이다
- 객체를 만들기 위한 설계도: Class
- 설계도로부터 만들어지는 개념: object
- 설계도로부터 실체화된(구현화된)객체 :instance

```python
class 집:
    # class 내부 요소 : 멤버(member)
    # 멤버 변수
    화장실 = 2
    방 = 4
    층 = 2
    다락방 = True
    # 멤버 함수
    def 정보(self):
        print(f'화장실 : {self.화장실}개')
        print(f'방 : {self.방}개')
        print(f'층 : {self.층}층')
        print(f'다락방 : {self.다락방}')
```

```python
예호집= 집()
예호집.방=5
예호집.화장실=1
예호집.정보()

print('----------------')

부모님집=집()
부모님집.정보()
```

```python
class 고양이:
    # 접근 제한자(Access modifier)
    
    
    # 좋아하는 사료, 무늬, 털색깔
    좋아하는사료= None
    무늬=None
    털색깔=None
    이름=None
    
    #울음소리, 먹이를 먹다
    def 울다(self):
        print('야옹')
    
    def 먹다(self):
        print('남냠')
        
    def 정보(self):
        print(f'이름:{self.이름}\n 좋아하는 사료:{self.좋아하는사료}\n 무늬:{self.무늬}\n 털색깔: {self.털색깔}')
        
# 생성자 (constructor)
# 객체가 생성될 때 자동으로 호출
    def __init__(self,이름,좋아하는사료,무늬,털색깔):
        self.이름=이름
        self.좋아하는사료= 좋아하는사료
        self.무늬=무늬
        self.털색깔=털색깔

```

```python
구름=고양이()
구름.이름='구름'
구름.좋아하는사료='애기사료'
구름.무늬='없음'
구름.털색깔='흰색'
```

```python
구름.정보()
```

```python
흰둥이=고양이('흰둥이', '츄르', '없음', '흰색')
흰둥이.정보()
흰둥이.이름='후추'
흰둥이.먹다()
흰둥이.울다()
```

### 추상화
- 객체의 공통적인 특징을 추출하는 작업
- 단, 너무 세부적인 정보 배제

### 상속(inheritance)
- 부모클래스의 멤버(변수, 함수)를 물려받아
- 자식클래스를 정의

### 캡슐화(Encapsulation)
- 개발자가 원하는 변수, 함수만 접근 할 수 있도록 
- 정보를 제한 

### 다형성(polymophism)
- 다양한 형태를 나타낼수 있는 능력

```python
class 샴고양이(고양이):
    털색깔= '투톤'
    
    def __init__(self, 이름, 좋아하는사료, 무늬):
        self.이름=이름
        self.좋아하는사료=좋아하는사료
        self.무늬=무늬
```

```python
샤오미=샴고양이('샤오미','츄르','없음')
샤오미.정보()
```

- 추상화: 공통적인 특징을 추출하여 만드는 과정
- 구체화: 자식클래스로 하나씩 만들어 가면서 구체화 시키는 과정

```python
class 계산기:
    
    def add(self, num1, num2):
        return num1+num2
    
    def add(self, num1, num2):
        return num1+num2
```

```python
계산기1=계산기()
계산기1.add(3,5)
```

```python
계산기1.add('3','5')
```

```python
계산기1.add(3,'5')
```
